```python
import psycopg2
conn = psycopg2.connect(host="postgres",database="osm", user="osm", password="osm")
cursor = conn.cursor()

streets = [
    'Redneck',
    'Hoes',
    'No Name',
    'Manlove',
    'Loveman',
    'Double Trouble',
    'Shades of Death',
    'Breakneck',
    'Purgatory',
    #'Wack Wack',
    'Weasel Drift',
    'La Rue',
    'Orphan Boy',
    'Gayboy',
    'Highboy',
    'Handyboy',
    'Ataboy',
    #'Neck Road',
    'Nummytown',
    'Boulevard Boulevard',
    'Sip Ave',
    'Ghost Pony',
    'Lovers Lane',
    'Unexpected Road',
    'Uncle Pete',
    'Butts Ave',
    'Buckshutem Road',
    'Big Enough Way'
]

results = []
for record in streets:
    cursor.execute("""
        select
         '%s' as name, ST_AsGeoJSON(ST_Transform(ST_UNION(way),4326)) as way
        FROM
           planet_osm_line
        WHERE
          lower(name) like '%%' || lower('%s') || '%%';
    """ % (record, record))
    for one in cursor:
        results.append(one)
    
cursor.close()
print (results)
```

```python
import json

geojson = {
  "type": "FeatureCollection",
  "features": []
}

for record in results:
    entry = {}
    entry['type'] = 'Feature'
    entry['geometry'] = json.loads(record[1])
    entry['properties'] = {'name': record[0]}
    geojson['features'].append(entry)

print (json.dumps(geojson, indent=2))
```
